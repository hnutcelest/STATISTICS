```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r"C:\Users\LENOVO\Desktop\stats\current-power.csv",delimiter=",")
df
```

```python
df.describe()
```

```python
df.cov()
```

```python
x=df['Current']
y=df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power')
plt.ylabel('Power')
plt.xlabel('Current')
plt.show()
```

```python
df.corr()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x,y, color='red' ,alpha=0.6,label='Data point')

plt.plot(x,line, color='blue' ,label=f'Regression line (r={r_value:.3f})')

plt.title('Device Current vs Power')
plt.ylabel('Power')
plt.xlabel('Current')
plt.legend()
plt.show
```

## Air Quality

```python
air = pd.read_csv(r"C:\Users\LENOVO\Desktop\air-quality.csv", delimiter=",")
air
```

# NOx

```python
x = air ['NOx']
y = air ['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x,y, color='magenta' ,alpha=0.6,label='Data point')

plt.plot(x,line, color='blue' ,label=f'Regression line (r={r_value:.3f})')

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show
```

# NO2

```python
x = air ['NO2']
y = air ['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x,y, color='red' ,alpha=0.6,label='Data point')

plt.plot(x,line, color='blue' ,label=f'Regression line (r={r_value:.3f})')

plt.title('NO2 vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NO2 (ug/cu.m)')
plt.legend()
plt.show
```
