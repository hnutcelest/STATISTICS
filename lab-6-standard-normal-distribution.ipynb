# Histogram

```python
plt.hist(df['Current'], bins=5,
         alpha=0.8,
         color="Magenta",
         edgecolor="Magenta")

plt.title('Histogram of Current Values',
          font = 'Palatino Linotype',
          fontsize = 12)

plt.xlabel('Current (mA)')
plt.ylabel('Frequency')
plt.show()
```

# Normal Distribution

```python
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
         color="red",
         linewidth=2.9)

plt.title('Normal Distribution of Current Values',
          font = 'Cambria',
          fontsize = 12)

plt.xlabel('Current (mA)')
plt.ylabel('Prbability Density')
plt.show()
```

```python
plt.hist(df['Current'], bins=5,
         density=True,
         alpha=0.8,
         color="blue",
         edgecolor="blue")

plt.title('Normal Distribution of Current Values',
          font = 'Cambria',
          fontsize = 12)

p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
         color="green",
         linewidth=2.9)
plt.show()
```

```python
x_min = df['Z-score'].min()
x_max = df['Z-score'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, df['Z-score'].mean(), df['Z-score'].std(ddof=1))

plt.plot(x, p,
         color="red",
         linewidth=2.9)

plt.title('Standard Normal Distribution of Current Values N(0, 1)',
          font = 'Cambria',
          fontsize = 12)

plt.xlabel('Z-score')
plt.ylabel('Prbability Density')
plt.show()
```
