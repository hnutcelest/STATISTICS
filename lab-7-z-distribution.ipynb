```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'transformer-voltage (1).csv',
               delimiter=",",
               index_col='Sample')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
alpha = 0.05
z_critical = stats.norm.ppf(1-alpha/2)
z_critical
```

```python
pop_std = 5

ci_lower, ci_upper = stats.norm.interval(
    confidence = 0.95,
    loc = df['Voltage'].mean(),
    scale = pop_std/np.sqrt(df['Voltage'].count())
)

print("CI =", "[",  ci_lower, ",", ci_upper,"]")
```

```python
mean = df['Voltage'].mean()
std =df['Voltage'].std(ddof=1)

x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
         color="green",
         linewidth=2.9)

plt.hist(df['Voltage'], bins=5,
         density=True,
         alpha=0.8,
         color="blue",
         edgecolor='white')

plt.title('Normal Distribution of Current Values',
          font = 'Cambria',
          fontsize = 12)
plt.ylabel('Probability Density')
plt.xlabel('Voltage (V)')

plt.axvline(x = ci_lower,
            color="Black",
            linestyle='--')
plt.axvline(x = ci_upper,
            color='Black',
            linestyle='--')
plt.show()
```
